```python
import emoji
import re

print("First Cleaning")

def to_lowercase(text):
    return text.lower()

def clean_text(text):
    text = str(text)

    # delete emoji
    text = emoji.replace_emoji(text, replace='')

    # delete URL
    text = re.sub(r'http\S+|www\S+', '', text)

    # delete mention
    text = re.sub(r'@\w+', '', text)

    # delete angka
    text = re.sub(r'\d+', '', text)

    # delete semua karakter non-alfabet
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # delete spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# slang
slang_dict = {
    'yg': 'yang',
    'dg': 'dengan',
    'dgn': 'dengan',
    'utk': 'untuk',
    'dpt': 'dapat',
    'kpd': 'kepada',
    'dr': 'dari',
    'drpd': 'daripada',
    'kalo': 'kalau',
    'klo': 'kalau',
    'kl': 'kalau',
    'kpn': 'kapan',
    'knp': 'kenapa',
    'kyk': 'seperti',
    'kayak': 'seperti',
    'krn': 'karena',
    'karna': 'karena',
    'sbg': 'sebagai',
    'ttg': 'tentang',
    'dlm': 'dalam',
    'pd': 'pada',
    'sm': 'sama',
    'sma': 'sama',
    'utg': 'utang',
    'utang2': 'utang-utang',
    'jln': 'jalan',
    'moga' : 'semoga',
    'smoga' : 'semoga',
    'smg' : 'semoga',
    'smga' : 'semoga',
    'jabat' : 'jabatan',
    'tri' : 'menteri',
    'bs' : 'bisa',
    'kalo' : 'kalau',
    'kl' : 'kalau',

    # kata Sifat
    'bgt': 'banget',
    'banget': 'sangat',
    'bgtu': 'begitu',
    'gt': 'begitu',
    'gitu': 'begitu',
    'gini': 'begini',
    'bener': 'benar',
    'beneran': 'benaran',
    'slalu': 'selalu',
    'skrg': 'sekarang',
    'skrng': 'sekarang',
    'byk': 'banyak',
    'bnyk': 'banyak',
    'lg': 'lagi',
    'trus': 'terus',
    'trs': 'terus',
    'aja': 'saja',
    'sja': 'saja',
    'cpt': 'cepat',
    'cepet': 'cepat',
    'msh': 'masih',
    'mshh': 'masih',

    # pronouns
    'gw': 'saya',
    'gua': 'saya',
    'sy': 'saya',
    'sa': 'saya',
    'lu': 'kamu',
    'elo': 'kamu',
    'km': 'kamu',
    'kmu': 'kamu',
    'anda': 'anda',
    'doi': 'dia',
    'mrk': 'mereka',
    'dya': 'dia',

    # kata kerja
    'blg': 'bilang',
    'blng': 'bilang',
    'liat': 'lihat',
    'liat2': 'lihat-lihat',
    'blm': 'belum',
    'sdh': 'sudah',
    'udh': 'sudah',
    'udah': 'sudah',
    'bs': 'bisa',
    'bsa': 'bisa',
    'byr': 'bayar',
    'jd': 'jadi',
    'jdi': 'jadi',
    'brp': 'berapa',
    'ngomong': 'berbicara',
    'ngmg': 'berbicara',

    # negasi
    'g': 'tidak',
    'ga': 'tidak',
    'gak': 'tidak',
    'gk': 'tidak',
    'enggak': 'tidak',
    'nggak': 'tidak',
    'tdk': 'tidak',
    'bkn': 'bukan',
    'bknnya': 'bukannya',
    'jgn': 'jangan',
    'jgn2': 'jangan-jangan',

    # konjungsi
    'tp': 'tapi',
    'tpi': 'tapi',
    'pdhl': 'padahal',
    'sbnrnya': 'sebenarnya',
    'emg': 'memang',
    'sblm': 'sebelum',
    'stlh': 'setelah',
    'moga': 'semoga',
    'dlu': 'dulu',
    'dulu': 'dulu',

    # topik
    'menkeu': 'menteri keuangan',
    'kemenkeu': 'kementerian keuangan',
    'pak': 'bapak',
    'bpk': 'bapak',
    'bu': 'ibu',
    'dpr': 'dewan perwakilan rakyat',
    'pem': 'pemerintah',
    'pemrth': 'pemerintah',
    'pemerntah': 'pemerintah',
    'ri': 'republik indonesia',
    'ind': 'indonesia',
    'indo': 'indonesia',
    'pres': 'presiden',
    'wapres': 'wakil presiden',
    'ekon': 'ekonomi',
    'pln': 'perusahaan listrik negara',
    'bbm': 'bahan bakar minyak',
    'subsdi': 'subsidi',
    'pjk': 'pajak',
    'ppn': 'pajak pertambahan nilai',
    'org': 'orang',

    # kata kerja ga perlu
    'wkwk': '',
    'wkwkwk': '',
    'haha': '',
    'hehe': '',
    'lol': '',
    'cuy': '',
    'bro': '',
    'sist': '',
    'gan': '',
    'guys': '',
    'sih': '',
    'dong': '',
    'deh': '',
    'kok': '',
    'lho': '',
    'loh': '',
    'woww': 'wow',
}

def normalize_slang(text):
    words = text.split()
    normalized_words = [slang_dict[word] if word in slang_dict else word for word in words]
    return ' '.join(normalized_words)

print("Semua fungsi 'General Cleansing' siap.")
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
import re

print("Menyiapkan tools untuk K-Means (Stopword & Stemmer)...")

factory_stem = StemmerFactory()
stemmer = factory_stem.create_stemmer()

factory_stop = StopWordRemoverFactory()
stopwords_bawaan = factory_stop.get_stop_words()

custom_stopwords = [
    'nya', 'yg', 'yang', 'di', 'ke', 'dari', 'ini', 'itu', 'dan', 'atau',
    'buat', 'sama', 'kalau', 'kalo', 'kl', 'ada', 'jadi', 'jdi', 'bisa', 'bs',
    'akan', 'pada', 'sudah', 'sdh', 'aja', 'saja', 'kok', 'sih', 'dong',
    'banget', 'bgt', 'karena', 'krn', 'apa', 'tapi', 'tp', 'mau', 'masih',
    'moga', 'smoga', 'biar', 'biarin', 'dah', 'deh', 'mah', 'tuh', 'ak', 'aku',
    'gw', 'gue', 'saya', 'lu', 'lo', 'kamu', 'anda', 'kita', 'kami', 'kalian',
    'banyak', 'dikit', 'semua', 'segala', 'kapan', 'dimana', 'siapa', 'bagaimana',
    'tak', 'tdk', 'nggak', 'ga', 'gak', 'bukan', 'donk', 'buk', 'ibuk',
    'bapak', 'pak', 'bpk', 'purbaya', 'sadewa', 'menteri', 'keuangan', 'menkeu', 'tri',
    'indonesia', 'negara', 'rakyat', 'orang', 'org', 'jabat', 'pejabat', 'sosok'
]

stopwords_lengkap = stopwords_bawaan + custom_stopwords

dictionary_stop = ArrayDictionary(stopwords_lengkap)
stopword_remover_fix = StopWordRemover(dictionary_stop)

def remove_stopwords(text):
    text = str(text)
    return stopword_remover_fix.remove(text)

def stem_text(text):
    return stemmer.stem(text)

print("Tools K-Means (Super Stopword Removal) siap.")
```

```python
print("Memulai pipeline preprocessing...")

df['text_display'] = df['text_display'].fillna('')

print("Tahap 1: Menerapkan General Cleansing (Case Folding, Cleaning, Normalization)...")

df['cleaned_text'] = df['text_display'].apply(to_lowercase)
df['cleaned_text'] = df['cleaned_text'].apply(clean_text)
df['cleaned_text'] = df['cleaned_text'].apply(normalize_slang)
print("Tahap 1 selesai.")

print("Tahap 2: Menerapkan pembersihan khusus K-Means (Stopword Removal & Stemming)...")
df['stemmed_text'] = df['cleaned_text'].apply(remove_stopwords)
df['stemmed_text'] = df['stemmed_text'].apply(stem_text)
print("Tahap 2 selesai.")

print("\nPreprocessing selesai. Menampilkan contoh perbandingan:")

display(df[['text_display', 'cleaned_text', 'stemmed_text']].head())
```
